##Create a cost function

Let's define all the current draws and timescales:

```python
# milliAmp current draw values, time durations in seconds
quiescent_current = 1.5 # mA
bluetooth_current = 20. # mA
bluetooth_duration = 10. # s
wifi_wakeup_current = 200. # mA
wifi_assoc_current = 60. # mA
wifi_powersave_current = 7. # mA
wifi_powersave_duration = 2*60 # s (2 minutes)
wifi_wakeup_duration = 0.5 # s (100-1000ms)
wifi_assoc_duration = 0.5 # s (100-1000ms)
locking_spike_current = 1500. # mA
locking_motion_current = 500. # mA
locking_spike_duration = 0.010 # s
locking_motion_duration = 2.5 # s

monitor_interval = 30*60. # s (30 minutes)
```

Let's find the power consumption of common events.
Assumption: wifi, motor, etc. are operating at a fixed voltage, so from P=IV => I_draw ~ P_consumed and Energy = Power * time

```python
wifi_no_command_event = wifi_wakeup_current*wifi_wakeup_duration + wifi_assoc_current*wifi_assoc_duration
wifi_command_event = wifi_no_command_event + wifi_powersave_current*wifi_powersave_duration
locking_event = locking_spike_current*locking_spike_duration + locking_motion_current*locking_motion_duration
```

Let's look at a day's worth of events (in milliamp hours):

```python
quiescent_current * one_day_in_seconds / one_hour_in_seconds
```

```python
wifi_no_command_event * (one_day_in_seconds / monitor_interval) / one_hour_in_seconds
```

```python
wifi_command_event * 5 / one_hour_in_seconds
```

```python
locking_event * 5 / one_hour_in_seconds
```

```python
def get_fitted_profile(user_activity_profile, global_prior=Y_scaled_global_totals, global_prior_weight=global_prior_weight, features=xx2):
    Y_user = global_prior * global_prior_weight + user_activity_profile
    treemodel.fit(features,Y_user)
    user_profile = treemodel.predict(features)
    return user_profile/user_profile.max()
```

```python
def plot_fitted_profile(user_activity_profile, global_prior=Y_scaled_global_totals, global_prior_weight=global_prior_weight, features=xx2, include_data=False):
    user_profile = get_fitted_profile(user_activity_profile, global_prior, global_prior_weight, features)
    plot(user_profile)
    ylim((0,1))
    if include_data:
        plot(user_activity_profile)
```

To approximate the number of pings, we calculate: $N_{pings} \approx \int \text{freq}(t) dt = \text{sum(freq_profile_values)*bin_time} = \text{sum(profile_values)*max_freq*bin_time}$
